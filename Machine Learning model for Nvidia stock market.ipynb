```python

# Example: assume df is your DataFrame and you want to subtract 'High' - 'Low'
df['Price_Range'] = df['High'] - df['Low']
```

```python
df.head()
```

```python
df['Price_Range']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['High','Low','Open','Volume','Price_Range']]
y = df['Close']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
lm.coef_
```

```python
edf = pd.DataFrame(lm.coef_, x.columns, columns=['coef'])
print(edf)
```

```python
predictions = lm.predict(x_test)
```

```python
predictions
```

```python
sns.scatterplot(x=predictions, y=y_test)
plt.xlabel('Predictions')
plt.title('Evaluation of our lm model')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', mean_squared_error(y_test, predictions))
print('AXSE:', math.sqrt(mean_absolute_error(y_test, predictions)))
```

```python
residuals = y_test - predictions
```

```python
residuals
```

```python
sns.displot(residuals, bins=50, kde=True)
```
